```python
import pandas as pd
pd.set_option('display.max_columns', 500)

sleeper_data = "sleeper_data.csv"
string_cols = ["location","name","race","lvl","physical","pierce","fire","electric","ice","force","mind","item"
               ,"physical_attack", "pierce_attack", "fire_attack", "electric_attack", "ice_attack", "force_attack"
               , "mind_attack"]
na_replace = "-"

df = pd.read_csv(
    sleeper_data
    ,dtype = {
        'location': str
        , 'name': str
        , 'race': str
        , 'lvl': str
        , 'physical': str
        , 'pierce': str
        , 'electric': str
        , 'ice': str
        , 'force': str
        , 'mind': str
        , 'item': str
        , 'physical_attack': str
        , 'pierce_attack': str
        , 'fire_attack': str
        , 'electric_attack': str
        , 'ice_attack': str
        , 'force_attack': str
        , 'mind_attack': str
    }
)
df[string_cols] = df[string_cols].fillna(na_replace)

df.head()
```

# Types

```python
types = ['physical', 'pierce', 'fire', 'electric', 'ice', 'force', 'mind']

location = "Hell"
df[df['location'] == location][['location', 'name'] + types]
```

# Attacks

```python
attacks = ['physical_attack', 'pierce_attack', 'fire_attack', 'electric_attack', 'ice_attack', 'force_attack'
           , 'mind_attack']

df[df['location'] == location][['location', 'name'] + attacks]
```
